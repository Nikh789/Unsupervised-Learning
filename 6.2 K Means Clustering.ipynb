```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
```

```python
univ = pd.read_csv("C:\\Users\\nikhi\\OneDrive\\Desktop\\Social Prachar\\Excel files\\Universities.csv")
univ
```

```python
del univ['Univ']
univ
```

```python
df_norm = (univ-univ.min())/(univ.max()-univ.min())
df_norm
```

```python
from sklearn.cluster import KMeans
wcss = [] # Within cluster-sum-of-squares
# Code to get max no of clusters
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state=0)
    kmeans.fit(df_norm)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11), wcss)
plt.title('The Elbow Method')
plt.xlabel('No of Clusters')
plt.ylabel('wcss')
```

```python
# Always cosider k values ODD
# There is a curve elbow at k = 3

import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(df_norm, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distance')
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters = 3, init = 'k-means++', random_state=0)
y_kmeans = model.fit_predict(df_norm)
y_kmeans
```

```python
X = np.array(df_norm)
```

```python
plt.scatter(X[y_kmeans==0,0], X[y_kmeans==0,1], s=100, c='red', label='Cluster1')
plt.scatter(X[y_kmeans==1,0], X[y_kmeans==1,1], s=100, c='blue', label='Cluster2')
plt.scatter(X[y_kmeans==2,0], X[y_kmeans==2,1], s=100, c='green', label='Cluster3')
```
