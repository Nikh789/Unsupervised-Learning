```python
del univ['Univ']
univ
```

```python
df_norm = (univ-univ.min())/(univ.max()-univ.min())
df_norm
```

```python
# Using Dendogram to find the optimal no of clusters
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(df_norm, method = 'ward'))
plt.title('Dendogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
# Fitting hierarchical clustering to the dataset
from sklearn.cluster import AgglomerativeClustering
```

```python
cluster = AgglomerativeClustering(n_clusters = 3, affinity = 'euclidean', linkage = 'complete')
```

```python
abc = cluster.fit_predict(df_norm)
# In case there is output variable 
# X_train, X_test, y_train, y_test
# Cluster.fit(X_train, y_train)
# Cluster.predict(X_test)
abc
```
